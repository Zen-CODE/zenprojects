# CAMI Strategy

## Introduction

CAMI currently faces a large challenge. We have a vast body of valuable, expensive content nested firmly in a aging technology. Supporting and developing this content is becoming increasingly difficult, both in terms of it's technical complexity and in terms of finding developers with the require skill set i.e. Visual Basic 6.

Although great strides have been made in terms of moving forward with more modern technologies, the content ported to far represents a small fraction of the total content. Not only that, some of the ported content does not meet the high standard of quality set by the Vb6 content. Most notably, some of the Perceptual exercises present only a superficial reflection of the original exercise conceptually, utilize low quality graphics and do not have audio. Then there is the vast body of the Reader that has not even started to be ported.

Given the limited human resources available to port this content, it seems unrealistic to suggest that CAMI could offer any equivalent replacement for the many years to come. How then, can we manage the migration of this content to newer platforms whilst still offering a fresh, appealing releases of the legacy EduSuite systems?

This document aims to provide a rough guide as to how this may be accomplished. It tries to paint a realistic picture of how CAMI development can be planned such that we:

- are able to offer existing clients an improved EduSuite each year.
- are able to produce content that can be deployed on both old and new technologies.
- can establish a smooth and natural path from the EduSuite to more modern web, mobile and native desktop technologies.
- can minimize the development effort in supporting older technologies.

## Phases

Such an effort will no doubt take a considerable time, and require careful consideration if we are to offer both an increasingly attractive offering to our clients, whilst allowing our development efforts to focus on utilizing newer technologies to accomplish this migration.

*Note*. The phases discussed here will focus more on clarifying the technical tasks than the specific content involved. The question of content is more detailed and complex, and probably best guided by requirements from clients or advice from content specialists.

## Phase 1 - Release 2020

**Objectives**

- offer the first Reader improvements using exercises utilizing portable content (Focus exercises).
- establish a functional web product offering Grade R to 7 Maths.
- update CAMI-App's to include English and Afrikaans spelling tests (to facilitate online/schools competitions and events)

**Milestones**

- New Reader spelling tests defined in Focus XML, included and functional for the 2020 Release.
- *CAMI Guanaco* offering web-based access CAMI Maths Grade R-7 (and whipping the Llamas ass!).
- CAMI-Apps Version 3, including English and Afrikaans Spelling tests.
- The CAMI-Apps server updated to include open Trophies for the above mentioned Spelling tests.

**Moonshots**

- Rework and improve lower quality Perceptual exercises.
- Port letter and number tracing exercises to XML (to feature in both Reader and Perceptual products).

- offer universal login and scoring capabilities across all CAMI products.
