# GAP Skills and Tools

GAP stands for "Growing As a Person". By applying some of the tricks outlined below, we can improve our ability to handle challenging situations and navigate difficult paths by improving our ability to make better decisions. 

## Mindfullness Skills

Mindfullness skills are based on the principle of managing out *Emotional mind* and *Rational mind*. These two aspects of our personalities are essential aspects of our personalities, but are also self interested and short sighted. They need to be unified and guided by the *Wise Mind*.

When we are overly controlled by either the rational or emotional mind, we can engage wise mind by stepping back mentally and:

* Observe - Step back. You are not your thoughts. Try and adopt a 3rd party perspective. Don't get swept away.
* Describe - Use wrods to distance yourelf from reacting. A thought is not a fact. Do not judge but descrive. e.g. replace "I am stupid" with "I acted stupidly".
* Participate - Enter wholly into the situation. Be spontaneous. Forget yourself. 

This approach helps you find your wise mind. How? The above steps help you be:

* Non-judgemental. Don' valuate. Unglue opinions. Accept. Acknowlege.
* Mindful. Do one thing at a time.  Let go of distraction. Focus.
* Effective. Focus on what works. Be skillful. Obey the rules. Focus on valued objectives.

## Distress Tolerance

Coping with distressful or difficult situation can be a challenge. If we are not guided by wise mind or th ehigher self, we can often act out in ways that sabotages our goals and the ideal outcome. Some useful mechanisms include:

* Self soothing
  - Engage the senses: vision, hearing, smell, touch, taste.
* Distract
  - Activities, hobbies, passtimes, disengaging, counter thoughts.
* Improve the moment
  - With imagery. A visualization that relaxes, motivates or comforts.
  - With meaning. Finding a purpose, meaning of value in the pain.
  - With relaxation. Exercise. Listen to music. Take a bath or have a coffee.
* One thing in the moment
  - Focus on the now. Breathe. Be aware of bodily sensations. 
* Think of the pros and cons

## Accepting Reality

The basic principles of *radical acceptance* invlove the deep acceptrance of *what is*. When we accept and ackowledge the pain, we can own it and deal with it.

*Turning the mind* is an act of choice and involves redirecting the mind awayh from rejecting reality towards acceptance. One has to make an inner commitment to accept, over and over again.

Points to remember that help in this process include recognizing that:

* Emotions are neither good not bad, right of wrong. They simply reflect the way we feel.
* Experiencing an emotion does not mean we have to act on it.
* Emotions don't last forever. They do pass.
* Emotions are not facts.
* Emotions serve important survival functions. Be willling to accept them.

## Interpersonal Effectiveness Skills

There are a few useful strategies to keep in mind when handling interpersonal interactions. We need to keep these in mind when

* Attending to relationships
* Balancing priorities vs. demands
* Balancing the *wants* to *shoulds* ratio
* Building mastery and self-respect

There are 3 ways to think more clearly and orient your wise mind in every conflict situation. Identify which of the following is most important:

* The Objective. Skills for objective effectiveness can be summarised as **DEAR MAN**.
    - Describe
    - Express
    - Assert
    - Reinforce
    - (stay) Mindful
    - Appear confident
    - Negotiate
* The Relationship. Maintaining or improving your relationship whilst tying to get what you want? **GIVE**
    - Gentle
    - Interested
    - Validate
    - Easy manner
* Your self-repect and integrity. **FAST**
    - Fair
    - (no) Apologies
    - Stick to value
    - Truthful
